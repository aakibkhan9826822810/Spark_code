Write a solution to show the unique ID of each user, If a user does not have a unique ID replace just show null.

```python
# Employees table
data = [
    (1,'Alice'),
    (7,'Bob'),
    (11,'Meir'),
    (90,'Winston'),
    (3,'Jonathan')
]

col = ['id','name']
employees  = spark.createDataFrame(data,col)
display(employees)
```

```python
data2 = [
    (3,1),
    (11,2),
    (90,3)
]

col2 = ['id','unique_id']

uni_id = spark.createDataFrame(data2,col2)
display(uni_id)
```

```python
result = (
    employees.join(uni_id,employees.id == uni_id.id, how='left')
    .select('unique_id','name')
    .orderBy('unique_id')
)
display(result)
```
