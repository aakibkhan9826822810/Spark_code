Write a solution to report the product_name, year, and price for each sale_id in the Sales table

```python
data = [
    (1,100,2008,10,5000),
    (2,100,2009,12,5000),
    (7,200,2011,15,9000)
]

col = ['sale_id','product_id','year','quantity','price']

sales = spark.createDataFrame(data, col)
display(sales)
```

```python

data2 = [
    (100,'Nokia'),
    (200,'Apple'),
    (300,'Samsung')
]

col2 = ['product_id','product_name']

product = spark.createDataFrame(data2, col2)
display(sales)
```

```python
p=product.alias('p')
s=sales.alias('s')
result = (
    p.join(s, on=p.product_id==s.product_id , how='inner')
    .select(p.product_name, s.year, s.price)
)
display(result)
```
