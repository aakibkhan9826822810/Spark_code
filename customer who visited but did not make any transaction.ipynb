Write a solution to find the IDs of the users who visited without making any transactions and the number of times they made these types of visits.

```python
data = [
    (1,23),
    (2,9),
    (4,30),
    (5,54),
    (6,96),
    (7,54),
    (8,54)
]

col = ['visit_id','customer_id']
customers = spark.createDataFrame(data, col)
customers.display()
```

```python

data2 = [
    (2,5,310),
    (3,5,300),
    (9,5,200),
    (12,1,910),
    (13,2,970)
]

col2 = ['transaction_id','visit_id','amount']
transactions = spark.createDataFrame(data2, col2)
transactions.display()
```

```python
from pyspark.sql.functions import col
c=customers.alias('c')
t=transactions.alias('t')

result = (
    c.join(t, on= c.visit_id == t.visit_id, how='left')
    .where(col('t.visit_id').isNull())
    .groupby('c.customer_id')
    .agg({'c.customer_id':'count'})
    .withColumnRenamed('count(c.customer_id)','count')
)
display(result)
```
