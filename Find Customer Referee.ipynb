Find the names of the customer that are either:

- referred by any customer with id != 2.
- not referred by any customer.
- Return the result table in any order.

The result format is in the following example.

```python
data = [
    (1,'Will',None),
    (2,'Jane',None),
    (3,'Alex',2),
    (4,'Bill',None),
    (5,'Zack',1),
    (6,'Mark',2)
]

col = ['id','name','referee_id']

customers = spark.createDataFrame(data,col)
display(customers)
```

```python
from pyspark.sql.functions import col

df = (customers.select('name')
      .where(col('referee_id').isNull() | (col('referee_id') != 2)
             ))
display(df)
```
