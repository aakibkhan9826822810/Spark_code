Write a solution to find all dates' id with higher temperatures compared to its previous dates (yesterday).

```python
from pyspark.sql.functions import col,cast

data = [
    (1,'2015-01-01',10),
    (2,'2015-01-02',25),
    (3,'2015-01-03',20),
    (4,'2015-01-04',30)
]

column = ['id','recordDate','temperature']
record = spark.createDataFrame(data, column)
df = record.withColumn('recordDate',col('recordDate').cast('date'))
display(df)
```

```python
from pyspark.sql.functions import col,date_add

d= df.alias('d')
t= df.alias('t')

result = (
    d.join(t, d.recordDate == date_add(t.recordDate, 1))
    .where(d.temperature > t.temperature)
    .select(d.id)
)
display(result)
```
