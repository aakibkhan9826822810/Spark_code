Write a solution to find the ids of products that are both low fat and recyclable.



```python

data = [
    (5, 'Y', 'N'),
    (1, 'Y', 'Y'),
    (2, 'N', 'Y'),
    (3, 'Y', 'Y'),
    (4, 'N', 'N')
]

col = ["product_id", "low_fats", "recyclable"]
products = spark.createDataFrame(data,col)
```

```python
from pyspark.sql.functions import col
df = products.select('product_id').where(col("low_fats") == "Y").where(col("recyclable") == "Y")
display(df)
```
